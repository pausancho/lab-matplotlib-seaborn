#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

avg_month_steps = fitbit.groupby(["Months_encoded"])["Steps"].mean().reset_index()
x_month = avg_month_steps["Months_encoded"]
y_steps = avg_month_steps["Steps"]
fig3, ax3 = plt.subplots(figsize=(12,4))
sns.barplot(x = x_month, y = y_steps, ax = ax3)
ax3.set_title("AVG Steps by Month")
#ax3.savefig("AVG Steps by Month.png")
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

steps = fitbit.groupby(["Work_or_Weekend"])["Steps"].mean().reset_index()
steps_week = fitbit["Steps"]
days = fitbit["Work_or_Weekend"]
fig4, ax4 = plt.subplots(figsize=(12, 4))
ax4.bar(days, steps_week)
ax4.set_title("AVG Steps by Week or Weekend")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

fitbit.head(5)
```

```python
x_col = "Steps"
y_columns = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]


for y_col in y_columns:

    figure = plt.figure
    ax5 = plt.gca()
    ax5.scatter(fitbit[x_col], fitbit[y_col])
    ax5.set_xlabel(x_col)
    ax5.set_ylabel(y_col)
    ax5.set_title("{} vs {}".format(y_col, x_col))

    plt.legend()
    plt.show()
```
